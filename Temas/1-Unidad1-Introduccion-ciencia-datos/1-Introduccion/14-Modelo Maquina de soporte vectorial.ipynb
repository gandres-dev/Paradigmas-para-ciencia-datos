### Importacion de nuestra librerias

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
```

### Cargando nuestos datos

```python
iris = datasets.load_iris()
X = iris.data[:, [0,2]]   # sepal length y petal length
y = iris.target 
```

Tanto el perceptron y la regresion lineal no funcionan tan bien con las columnas `[0,2]` en cambio, con esta maquina de soporte vectorial si lo hace.

### Modelo Maquina de soporte vectorial

```python
# C es un metaparametro, es algo simila a la eta del perceptron
# Kenel son nucleos de grupamiento
svm = SVC(C=0.5, kernel='linear')
svm.fit(X,y)   # directos sin pasar a la estandarizacion
```

### Graficas de nuestro modelo


```python
plot_decision_regions(X, y, clf=svm, legend=2)

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title("SVM para Iris")
plt.show()
```

## Comparacion con el perceptron

```python
from sklearn.linear_model import Perceptron

ppn = Perceptron(max_iter=40, eta0=0.1, random_state=1)
ppn.fit(X, y)
```

### Graficando el modelo

```python
plot_decision_regions(X, y, clf=ppn, legend=2)

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title("Perceptron para Iris")
plt.show()
```

No un cruze entre para las treas clases, solo hay una recta que separa nuestro conjunto de datos

## Comparacion con regrecesión logística

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=100.0, random_state=1)
lr.fit(X, y)
```

### Graficando nuestro modelo

```python
plot_decision_regions(X, y, clf=lr, legend=2)

plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title("Perceptron para Iris")
plt.show()
```

- Algo similar, pero la maquina soporte vectorial encuentra regiones.
