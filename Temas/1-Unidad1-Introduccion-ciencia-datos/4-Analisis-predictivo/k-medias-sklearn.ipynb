```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (12, 6)
plt.style.use('ggplot')
```

```python
data = pd.read_csv('http://dicyg.fi-c.unam.mx:8080/lalo/pypcd/presentaciones/xclara.csv')
print(data.shape)
data.head()
```

```python
x1 = data.V1.values
x2 = data.V2.values
X = np.array(list(zip(x1, x2)))
plt.scatter(x1, x2)
```

Se ve claramente que podemos agrupar el conjunto de datos.

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3)
km = km.fit(X)
y_hat = km.predict(X)
C = km.cluster_centers_
```

```python
fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:,1], c=y_hat)
ax.scatter(C[:,0], C[:,1], marker='*', s=200, c='k')
```
