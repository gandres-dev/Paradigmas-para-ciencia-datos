```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img_orig = cv2.imread('paisaje.jpg')
img = cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
img.shape
```

```python
vect = img.reshape((-1, 3))
vect = np.float32(vect)
criteria = (cv2.TermCriteria_MAX_ITER, 10, 1.10)
K = 5
attempts = 10
ret, label, center = cv2.kmeans(vect, K, None, criteria, attempts, cv2.KMEANS_PP_CENTERS)
```

```python
center = np.uint8(center)
res = center[label.flatten()]
img_res = res.reshape((img.shape))
```

```python
figure_size = 15
plt.figure(figsize=(figure_size, figure_size))
plt.subplot(1,2,1), plt.imshow(img)
plt.title('Original'), plt.xticks([]), plt.yticks([])
```

```python
figure_size = 15
plt.figure(figsize=(figure_size, figure_size))
plt.subplot(1,2,1), plt.imshow(img)
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2), plt.imshow(img_res)
plt.title('Imagen segmentada con k - %i' % K), plt.xticks([]), plt.yticks([])
plt.show()
```
