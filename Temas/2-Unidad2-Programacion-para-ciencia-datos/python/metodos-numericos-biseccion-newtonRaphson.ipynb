```python
#Metodo de biseccion
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

```python
def biseccion(f, a0, b0, emax=1e-5, iter_max=50, graf=True):
    """
    docstring here
    """
    a, b = a0, b0
    error, i = 10, 0
    C, Fc = [], []
    while error > emax and i < iter_max:
        c = (a+b) / 2  # mejor para buscar el a -b corrimiento
        fa = (a)
        fb = f(b)
        fc = f(c)
        C.append(c)
        Fc.append(fc)
        if fa * fc < 0:
            b = c
        else:
            a = c
        error = abs(fc)
        i += 1
    if graf: grafica(f, C, Fc)
    return i,c

def grafica(f, c, fc):
    mpl.style.use('seaborn')
    # Determinando el intervalo para la grafica
    cc = c.copy()
    cc.sort()
    ab = abs(cc[0] - cc[-1]) if (abs(cc[0] - c[-1]) > abs(cc[-1] - c[-1])) else abs(cc[-1] - c[-1])
    print(ab)
    x = np.linspace(c[-1] - ab, c[-1] + ab, 100)
    plt.plot(x, f(x), color='green')
    plt.scatter(c, fc, color='red', s=30)
    plt.scatter(c[-1], fc[-1], c='blue', s=40)
    plt.show()
```

```python
def f1(x):
    return x**3 - x**2 - 1
```

```python
x = np.linspace(-1, 3, 100)
plt.plot(x, f1(x))
```

```python
it, root = biseccion(f1, 1, 2)
```

```python
def f2(x):
    return np.sin(2 * x) * (x ** 3 - 2 * x ** 2 -1)
x = np.linspace(-4, 4, 100)
plt.plot(x, f2(x))
plt.show()
```

```python
biseccion(f2, -4, -3)
```

```python
# Derivada numerica
def Df(f, x, h=1e-3):
    return (f(x+h) - f(x)) / h

def f(x):
    return x ** 3 - 2 * x ** 2 -1

def f2(x):
    return np.sin(2 * x) * (x ** 3 - 2 * x ** 2 -1)
```

```python
x = np.linspace(0, 4 * np.pi, 100)
f = np.sin
fdy = np.cos # derivada simbolica

plt.plot(x, f(x), label='Original')
plt.plot(x, fdy(x), 'o', label='Simbolica')
plt.plot(x, Df(f,x), label='Numerica')

plt.legend()
plt.show()
```

```python
def newton_rapson(f, x0, emax=1e-5, iter_max=50, graf=True):
    error, i = 10, 0
    C, Fc = [], []
    while error > emax and i < iter_max:
        xi = x0 - f(x0) / Df(f, x0)
        C.append(x0)
        Fc.append(f(x0))
        error = abs(f(xi) - f(x0))
        x0 = xi
        i += 1
    if graf: grafica(f, C, Fc)
    return(i, x0)
```

```python
newton_rapson(f1, 1.6)
```

```python
newton_rapson(f2, -4)
```

```python
newton_rapson(f2, -3.5)
```

```python
newton_rapson(f2, -2)
```
